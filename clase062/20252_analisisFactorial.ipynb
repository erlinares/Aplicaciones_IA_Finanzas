## Análisis Factorial: Concepto y Utilidad

### Definición:
El análisis factorial es una técnica estadística multivariante que se utiliza para identificar relaciones subyacentes entre variables observadas. Su objetivo principal es reducir la dimensionalidad de los datos, agrupando variables correlacionadas en factores latentes no observables. Cada factor explica parte de la variabilidad total del conjunto de datos.

### Objetivos del Análisis Factorial:

* Reducción de Dimensionalidad: Simplificar un conjunto grande de variables en un número menor de factores.

* Identificación de Estructuras: Descubrir estructuras subyacentes en los datos.

* Eliminación de Redundancia: Eliminar redundancias en los datos y resaltar variables que proporcionan información única.

* Validación de Constructos: Verificar la validez de constructos teóricos en investigación.

### Supuestos del Análisis Factorial:

* Normalidad Multivariante: Las variables deben seguir una distribución normal multivariante.

* Linealidad: Las relaciones entre las variables deben ser lineales.

* Adecuación de la Muestra: El tamaño de la muestra debe ser adecuado para garantizar la estabilidad de los resultados (una regla común es tener al menos 5-10 observaciones por variable).

* Ausencia de Multicolinealidad Perfecta: No debe haber colinealidad perfecta entre las variables.

* Esfericidad: Las variables deben tener una correlación adecuada para realizar el análisis factorial (medido mediante el Test de Esfericidad de Bartlett y el Índice KMO).

```python
ev, v = fa.get_eigenvalues()
ev
plt.plot(range(1,data_af.shape[1]+1),ev)
```

```python
ev
```

```python
plt.figure(figsize=(8, 5))
plt.plot(range(1, data_af.shape[1]+1), ev, marker='o')
plt.xlabel('Número de Factores')
plt.ylabel('Eigenvalue')
plt.title('Scree Plot')
plt.grid(True)
plt.show()
```

En donde se muestra la caida grande ?

```python
fa = FactorAnalyzer(3, rotation='varimax')
fa.fit(data_af)
loads = fa.loadings_
print(loads)
```

Se observan los 3 posibles factores:

Comfort: Food and Drink, Seat comfort, Inflight entertainment, Cleanliness

Service: Onboard service, Baggage Handling, Inflight Service

Convenience: In flight Wifi, Departure/Arrival time convenience, Online Booking, Gate Location.

Corroborar si los factores son correctos utilizando Alfa de Cronbach

```python
pip install pingouin #Downloading pingouin-0.5.2.tar.gz (185 kB)
```

```python
import pingouin as pg
```

```python
#Creando los factores
factor1 = data[['Food and drink', 'Seat comfort', 'Inflight entertainment', 'Cleanliness']]
factor2 = data[['On-board service', 'Baggage handling', 'Inflight service']]
factor3 = data[['Inflight wifi service', 'Departure/Arrival time convenient', 'Ease of Online booking', 'Gate location']]

#obtener el alfa de cronbach
factor1_alpha = pg.cronbach_alpha(factor1)
factor2_alpha = pg.cronbach_alpha(factor2)
factor3_alpha = pg.cronbach_alpha(factor3)

print(factor1_alpha, factor2_alpha, factor3_alpha)
```

Los alfa Cronbach mayor a 0.7  son suficientes para garantizar la fiabilidad de la escala
