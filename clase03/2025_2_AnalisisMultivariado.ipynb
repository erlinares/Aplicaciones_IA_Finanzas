# Graficar

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pd.plotting.scatter_matrix(dataset.loc[:,"V2":"V14"], diagonal="kde", figsize=(20,15))
plt.show()
```

# Grafica 2

```python
import seaborn as sns
```

```python
dataset.V1.unique()
```

```python
for i in range(2,14):
  sns.lmplot(x="V"+str(i), y="V"+str(i + 1), data=dataset, hue="V1" , fit_reg=False)
```

# Matriz de correlación

```python
cormat = dataset.corr()
cormat
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(cormat, cmap='coolwarm' ,annot=True)
```
