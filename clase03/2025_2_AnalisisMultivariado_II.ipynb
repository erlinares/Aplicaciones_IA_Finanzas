# Ejercicio 1

Imagina que tienes los gastos mensuales de varias familias y quieres entender patrones de consumo.


```python
import numpy as np
import pandas as pd
import seaborn as sns
```

## Leer datos

```python
dataset = pd.read_csv("gastos_data.csv")
```

## Información del dataset

```python
dataset.info()
```

## Mostrar los primeros 5

```python
dataset.head(5)
```

## Mostrar los últimos 5

```python
dataset.tail(5)
```

## Mostrar la estadística descriptiva

```python
dataset.describe()
```

## Tamaño del dataset

```python
dataset.shape
```

## Histogramas

```python
dataset.Comida.hist()
```

## Graficar

```python
import matplotlib.pyplot as plt
%matplotlib inline
pd.plotting.scatter_matrix(dataset, diagonal="kde", figsize=(20,15))
plt.show()
```

## Matriz de correlación

```python
cormat = dataset.corr()
cormat
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(cormat, cmap='coolwarm' ,annot=True)
```
